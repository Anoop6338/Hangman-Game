```python
import pygame
import math
import random
import time
#import cv2
#import numpy as np
pygame.mixer.init()




pygame.init()
width, height = 800,500
win = pygame.display.set_mode((width,height))
pygame.display.set_caption("Hangman Game!")

#alphabets
radius = 20
gap = 15
letters = []
startx = round((width - (radius*2 + gap)*13)/2)
starty = 350
A = 65
for i in range(26):
    x = startx + gap*2 + ((radius*2 + gap)*(i%13))
    y = starty + ((i//13) * (gap + radius*2))
    letters.append([x,y, chr(A+i), True])

    
    
    
#load images
images = []
for i in range(7):
    image = pygame.image.load("hangman"+ str(i)+ ".png")
    images.append(image)

#background img
bg_img = pygame.image.load("img_bg03.jpeg")
bg_img = pygame.transform.scale(bg_img, (width,height)).convert_alpha()

#start bg
start_img =  pygame.image.load("hangman_bg03.jpeg")
start_img = pygame.transform.scale(start_img, (width,height)).convert_alpha()

#lose img
lose_img = pygame.image.load("lose_bg02.jpeg")
lose_img = pygame.transform.scale(lose_img, (800,500)).convert_alpha()


#win img
win_img = pygame.image.load("win_bg01.jpeg")
win_img = pygame.transform.scale(win_img, (width,height)).convert_alpha()



    
    
#game variable
hangman_status = 0
words = ["HELLO", "PYTHON", "HACK","LAPTOP","FOOTBALL","CRICKET","REDBULL","RING","JARVIS","CAP","HULK","MIKE","TURBAN","SPARK","ALEXA","SPOTIFY","BULB","CHAIR","TABLE","BLANKET","GRAPHICS","MAGGI","GREEN","COAT","TINDER","APPLE","CARTOON","PILLAR","REEBOK","CHANNEL","GUCCI","GENIUS","HACKATHON","WATCH","MOBILE","HEADPHONES","CARD","DELL","SHAKE","JEANS","CAMERA","OFFICE","BOTTLE","JAVA","GOOGLE","FARMING"]
word = random.choice(words)
guessed = []


#fonts
letter_font = pygame.font.SysFont('comicsans',40)
word_font = pygame.font.SysFont('comicsans',70)
title_font = pygame.font.SysFont('comicsans',90)




# colors
WHITE = (255,255,255)
BLACK = (0,0,0)



#game setup
FPS=60
clock = pygame.time.Clock()
run = True

#def start_draw():
    #win.blit(start_img,(0,0))
    #pygame.display.update()

def start():
    pygame.time.delay(100)
    pygame.mixer.music.load('welcome_bg.mp3')
    pygame.mixer.music.play()
    win.blit(start_img,(0,0))
    pygame.display.update()
    pygame.time.delay(2000)
    





pygame.mixer.music.load('main_bg01.mp3')
pygame.mixer.music.play()
#pygame.mixer.music.load('BG01.mp3')
#pygame.mixer.music.play()
def draw():
    #main game
    #pygame.mixer.music.load('main_bg01.mp3')
    #pygame.mixer.music.play()
    win.blit(bg_img,(0,0))
    
    #draw title
    text = title_font.render("HANGMAN",1,BLACK)
    win.blit(text,(width/2 - text.get_width()/2, 6))
   
    #drawing words
    display_word = ""
    for letter in word:
        if letter in guessed:
            display_word += letter + " "
        else:
            display_word += "_ "
    text = word_font.render(display_word, 1, BLACK)
    win.blit(text,(300,200))
    
    #buttons
    for letter in letters:
        x, y, ltr, visible = letter
        if visible:
            pygame.draw.circle(win,BLACK, (x, y), radius,3)
            text = letter_font.render(ltr,1, BLACK)
            win.blit(text, (x- text.get_width()/2,  y-text.get_height()/2))

    win.blit(images[hangman_status],(100,70))
    pygame.display.update()
    
    
def display_mssg(message):
    pygame.time.delay(1000)
    #win.fill(WHITE)
    
    text = word_font.render(message ,1,BLACK)
    win.blit(text, (width/2 - text.get_width()/2,  height/2 - text.get_height()/2))
    pygame.display.update()
    pygame.time.delay(3000)
    
 

while run:
    start()
    break

while run:
    
    clock.tick(FPS)
    
    
    draw()  
    
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            run = False
        
        if event.type == pygame.MOUSEBUTTONDOWN:
            m_x, m_y = pygame.mouse.get_pos()
            for letter in letters:
                x,y ,ltr, visible= letter
                if visible:
                    dis = math.sqrt((x-m_x)**2 + (y-m_y)**2)
                    if dis<radius:
                        letter[3]=False
                        guessed.append(ltr)
                        if ltr not in word:
                            hangman_status +=1
    
    draw()
                            
    won = True                   
    for letter in word:
        if letter not in guessed:
            won = False
            break
    
    if won:
        win.blit(win_img,(0,0))
        pygame.mixer.music.load('win_bg.mp3')
        pygame.mixer.music.play()
        #display_mssg("YOU WIN!!")
        pygame.display.update()
        time.sleep(2)
        break
    
    if hangman_status == 6:
        time.sleep(2)
        #display_mssg("YOU LOST!!")
        pygame.mixer.music.load('lose_bg.mp3')
        pygame.mixer.music.play()
        win.blit(lose_img,(0,0))
        pygame.display.update()
        time.sleep(2)
        break

            
pygame.quit()


            
```
